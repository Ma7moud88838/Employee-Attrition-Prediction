```python
df_train = pd.read_csv("train.csv")
df_train
```

```python
df_test = pd.read_csv("test.csv")
df_test
```

```python
df = pd.concat([df_train, df_test], ignore_index=True)
df
```

```python
df.to_csv("attrition_data.csv")
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.drop("Employee ID", axis = 1, inplace=True)
```

```python
df.describe()
```

```python
cols = ["Years at Company", "Monthly Income", "Monthly Income", "Company Tenure"]
for col in cols:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  df = df[(df[col] >= (Q1 - 1.5 * IQR)) & (df[col] <= (Q3 + 1.5 * IQR))]

df.info()
```

```python
df.describe()
```

```python
cols = list(df.columns)
for i in range(len(cols)):
  cols[i] = cols[i].replace(" ", '_')
  cols[i] = cols[i].replace("-", '_')
df.columns = cols
df.columns
```

```python
df.head()
```

```python
for col in df.columns:
  if(df[col].dtype == object):
    print(df[col].value_counts())
    print("_______________________________")
```

```python
encoding_map = {'Male':1, 'Female':0, 'Yes':1, 'No':0, 'Poor':1, 'Fair':2, 'Good':3, 'Excellent':4, 'Low':1, 'Medium':2, 'High':3, 'Very High':4, 'Low':0, 'Below Average':1, 'Average':2, 'High School':1, 'Associate Degree':2, 'Bachelor’s Degree':3, 'Master’s Degree':4, 'PhD':5, 'Entry':1, 'Mid':2, 'Senior':3, 'Small':1, 'Large':3, 'Stayed':1, 'Left':0}
for col in ['Gender', 'Work_Life_Balance', 'Job_Satisfaction', 'Performance_Rating',
       'Overtime', 'Education_Level', 'Job_Level', 'Company_Size', 'Remote_Work',
       'Leadership_Opportunities', 'Innovation_Opportunities',
       'Company_Reputation', 'Employee_Recognition', 'Attrition']:
  df[col] = df[col].map(encoding_map)

df
```

```python
df.info()
```

```python
df_encoded = pd.get_dummies(df, columns=["Marital_Status", "Job_Role"])
df_encoded.head()
```

```python
df_encoded.drop(["Job_Role_Technology", "Marital_Status_Single"], axis = 1, inplace=True)
df_encoded.info()
```

```python
for col in df_encoded.columns:
  if df_encoded[col].dtype == bool:
    df_encoded[col] = df_encoded[col].astype(np.int64)

df_encoded
```

```python
df_encoded.info()
```

```python

X_tot = np.array(df_encoded.drop("Attrition", axis = 1))
y_tot = np.array(df_encoded["Attrition"])
print(X_tot.shape)
print(y_tot.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X_tot, y_tot, test_size=0.2, random_state=42, shuffle=True)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import pickle

# save the iris classification model as a pickle file
model_pkl_file = "Standard_Scaler.pkl"

with open(model_pkl_file, 'wb') as file:
    pickle.dump(scaler, file)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

# Classification Report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

# Classification Report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(11)
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

# Classification Report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=9)
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

# Classification Report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(max_depth=12)
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

# Classification Report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier()
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

# Classification Report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

```python


# save the iris classification model as a pickle file
model_pkl_file = "GradientBoostingClassifier.pkl"

with open(model_pkl_file, 'wb') as file:
    pickle.dump(model, file)
```
